```python
def combine_dics(diclist):
    dic_combined = {}
    for dic in diclist:
        for key, val in dic.iteritems():
            if not key in dic_combined:
                dic_combined[key] = val
            else:
                dic_combined[key] = dic_combined[key] + val
    return dic_combined

data = combine_dics(diclist_0328_0613)
```

```python
day_counts = {}
for turnstile, rows in datetime_counts.items():
    by_day = {}
    for time, count in rows:
        day = time.date()
        by_day[day] = by_day.get(day, 0) + count
    day_counts[turnstile] = sorted(by_day.items())
```

```python
day_counts.items()[0]
```

```python
dictlist =[]
for key, value in day_counts.iteritems():
    vallist = []
    for k, v in value:
        val = [k,v]
        vallist.append(val)
    temp = [key, vallist]
    dictlist.append(temp)
    
dictlist[0]
```

```python
data_station = {}
data_station = defaultdict(list)
for l in dictlist:
    key = l[0]
    value = l[1]
    data_station[key].append(value)
    
sorted(data_station.items())[0]
```

```python
for key, value in data_station.iteritems():
    dd = defaultdict(int)
    for index, val in enumerate(value):
        for i, v in enumerate(val):
            date = v[0]
            dd[date] += v[1]
    data_station[key] = dd
```

```python
Timesq = defaultdict(list)
for key, value in data_station.iteritems():
    if key[3] == '42 ST-TIMES SQ':
        for k, v in value.iteritems():
            tempkey = k
            val = v
            Timesq[tempkey].append(val)
```

```python
TSday = defaultdict(int)
for key, value in Timesq.iteritems():
    for i, val in enumerate(value):
        date = key
        TSday[date] += val
```

```python
TSlist = []
for key, value in TSday.iteritems():
    temp = (key, value)
    TSlist.append(temp)
```

```python
TSdates, TScounts = zip(*sorted(TSlist))
plt.figure(figsize=(10,3))
plt.plot(TSdates,TScounts)
plt.ylabel('Total Traffic')
```
